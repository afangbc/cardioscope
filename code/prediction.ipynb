```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
rawdata = pd.read_csv("CVD Dataset.csv")
rawdata.head()
```

```python
rawdata['CVD Risk Score'].hist()
```

```python
rawdata.isna().sum()
```

```python
rawdata.head()
```

```python
rawdata.columns
```

```python
rawdata['BMI'] = rawdata['Weight (kg)']/(rawdata['Height (m)']**2)
```

```python
rawdata['Height (cm)'] = rawdata['Height (m)'] * 100
```

```python
rawdata['Height (m)'] = rawdata['Height (cm)'] / 100
```

```python
rawdata.head()
```

```python
rawdata.dropna(inplace=True)
```

```python
rawdata.shape
```

```python
rawdata.select_dtypes('object')
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
label_encoder = LabelEncoder()
one_hot_encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
```

```python
rawdata['Blood Pressure Category'].unique()
```

```python
rawdata['CVD Risk Level'].unique()
```

```python
rawdata['Smoking Status']= label_encoder.fit_transform(rawdata['Smoking Status'])
rawdata['Diabetes Status']= label_encoder.fit_transform(rawdata['Diabetes Status'])
rawdata['Sex']= label_encoder.fit_transform(rawdata['Sex'])
rawdata['Family History of CVD']= label_encoder.fit_transform(rawdata['Family History of CVD'])
```

```python
map = {
    'Physical Activity Level': ['Low', 'Moderate', 'High'],
    'Blood Pressure Category': ['Normal', 'Elevated', 'Hypertension Stage 1', 'Hypertension Stage 2']
}
```

```python
for col, order in map.items():
    mapping = {label: i for i, label in enumerate(order)}
    rawdata[col] = rawdata[col].map(mapping)
```

```python
rawdata.head()
```

```python
rawdata.drop(['Blood Pressure (mmHg)', 'CVD Risk Level'], axis=1, inplace=True)
```

```python
rawdata.head()
```

```python
plt.figure(figsize=(14, 10))
sns.heatmap(rawdata.corr().round(2), annot=True,annot_kws={"size": 10})
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
rawdata['ratio'] = rawdata['Estimated LDL (mg/dL)'] / rawdata['HDL (mg/dL)']
rawdata['ratio'].hist()
```

```python
from sklearn.model_selection import train_test_split
```

```python
# features = ['Sex', 'Age', 'Weight (kg)', 'Height (m)', 'BMI',
#       'Abdominal Circumference (cm)', 'Total Cholesterol (mg/dL)',
#       'HDL (mg/dL)', 'Fasting Blood Sugar (mg/dL)', 'Physical Activity Level',
#       'Family History of CVD', 'CVD Risk Level', 'Height (cm)',
#       'Waist-to-Height Ratio', 'Systolic BP', 'Diastolic BP',
#       'Blood Pressure Category', 'Estimated LDL (mg/dL)',
#       'Smokes', 'Diabetes', 'Excercises']
# Importances according to RandomForest (importance above 0.05)
# features = ['Weight (kg)','BMI', 'Systolic BP','Total Cholesterol (mg/dL)','Estimated LDL (mg/dL)','Diabetes']
# Importances according to heatmap
features = ['Diabetes Status','Estimated LDL (mg/dL)','Systolic BP','Total Cholesterol (mg/dL)','BMI', 'Weight (kg)']

X = rawdata[features]
y = rawdata['CVD Risk Score']
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=1000)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
# Model 1: Linear Regression
from sklearn.linear_model import LinearRegression
linear_regression_model = LinearRegression().fit(X_train, y_train)
linear_regression_predictions = linear_regression_model.predict(X_test)
lr_mae = np.sqrt(mean_absolute_error(linear_regression_predictions, y_test))
print(lr_mae)
```

```python
# Model 2: Decision Tree
from sklearn.tree import DecisionTreeRegressor
tree_model = DecisionTreeRegressor(max_depth=8, min_samples_leaf=5).fit(X_train, y_train)
tree_predictions = tree_model.predict(X_test)
dt_mae = np.sqrt(mean_absolute_error(tree_predictions, y_test))
print(dt_mae)
```

```python
# Model 3: Random Forest
from sklearn.ensemble import RandomForestRegressor
forest_model = RandomForestRegressor(n_estimators=100,min_samples_leaf=5).fit(X_train, y_train)
forest_predictions = forest_model.predict(X_test)
rf_mae = np.sqrt(mean_absolute_error(forest_predictions, y_test))

print(rf_mae)
```

```python
# Model 4: XGBoost Regressor
import xgboost as xgb
XGB_model = xgb.XGBRegressor().fit(X_train.to_numpy(), y_train)
XGB_predictions = XGB_model.predict(X_test.to_numpy())
xb_mae = np.sqrt(mean_absolute_error(XGB_predictions, y_test))
print(xb_mae)
```

```python
importances = forest_model.feature_importances_
importances_forest = pd.DataFrame({'Features':features, 'Importance Level':importances})
ax = importances_forest.plot.bar(x='Features', y='Importance Level',rot=90)
```

```python
import matplotlib.pyplot as plt

# Model names and RMSE values
models = ['Linear Regression', 'Decision Tree', 'Random Forest', 'XGBoost']
mae_values = [lr_mae, dt_mae, rf_mae, xb_mae]

# Create the bar plot
plt.figure(figsize=(8,5))
plt.bar(models, mae_values)
plt.xlabel('Models')
plt.ylabel('MAE')
plt.title('MAE Comparison of Models')
plt.ylim(0, max(mae_values) + 0.5)  # adds some space above bars

# Show the plot
plt.show()
```
